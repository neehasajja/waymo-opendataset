```python
import os
import pyarrow.parquet as pq
import tensorflow as tf

# Set the path to the directory containing the parquet files
dir_path = '/content/drive/MyDrive/testing/lidar_box/'
# Get a list of all the files in the directory
files = os.listdir(dir_path)

# Loop through each file and extract metadata
for filename in files:
    # Skip directories
    if os.path.isdir(os.path.join(dir_path, filename)):
        continue
        
    # Check if the file is a parquet file
    if filename.endswith('.parquet'):
        # Read the parquet file
        parquet_file = pq.ParquetFile(os.path.join(dir_path, filename))

        # Get the metadata
        metadata = parquet_file.metadata

        # Print the metadata
        print("File:", filename)
        print(metadata)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the Parquet file into a DataFrame
df = pd.read_parquet('/content/drive/MyDrive/training/lidar_box/file1.parquet')

# Select the columns you want to visualize
cols = ['feature1', 'feature2', 'target']

# Create a scatter plot of the data
plt.scatter(df[cols[0]], df[cols[1]], c=df[cols[2]], cmap='viridis')
plt.xlabel(cols[0])
plt.ylabel(cols[1])
plt.colorbar(label=cols[2])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the .parquet file
data = pd.read_parquet('/content/drive/MyDrive/training/lidar_box/file1.parquet')

# Select the columns to visualize
columns_to_visualize = ['[LiDARBoxComponent].box.size.x', '[LiDARBoxComponent].box.size.y']

# Group the data by object id and get the mean of the selected columns
grouped_data = data.groupby('key.laser_object_id')[columns_to_visualize].mean()

# Create a scatter plot of the grouped data
plt.scatter(grouped_data['[LiDARBoxComponent].box.size.x'], grouped_data['[LiDARBoxComponent].box.size.y'])

# Set the title and axis labels
plt.title('Average Box Size of Objects in LiDAR Data')
plt.xlabel('Box Size X')
plt.ylabel('Box Size Y')

# Display the plot
plt.show()
```
